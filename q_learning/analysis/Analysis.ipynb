# Importing Libraries
* pandas for data manipulation
* matplotlib for plotting graphs

```
import pandas as pd
import matplotlib.pyplot as plt
```

# Reading the scores file and epsilon values file

```
file_1 = 'scores.pkl'
file_2 = 'epsilons.pkl'

scores = pd.read_pickle(file_1)
epsilons = pd.read_pickle(file_2)
```

# Total reward received by the agent vs episodes

### The reward received in the first few episodes is less as the agent is exploring during this stage. Gradually as the number of episodes increases, the agent understands the game and takes better actions and hence the reward increases over time. However, the agent is not perfect and takes some bad decisions also which explains the uneveness of the curve

```
x_axis = [i for i in range(1, len(scores)+1)]

plt.plot(x_axis, scores) 
plt.xlabel("Episodes")
plt.ylabel("Total Reward")
plt.title("Reward received by the agent vs number of episodes")
plt.show()
```

# Epsilon Decay

### At the start the value of epsilon is 1 which means that the agent keeps exploring. However as the number of episodes increases the value of epsilon decays and the agent starts choosing greedy actions (exploitation) to maximize reward. The smallest value of epsilon is 0.001

```
x_axis = [i for i in range(1, len(scores)+1)]

plt.plot(x_axis, epsilons) 
plt.xlabel("Episodes")
plt.ylabel("Epsilon")

plt.title("Epsilon decay")
plt.show()
```
